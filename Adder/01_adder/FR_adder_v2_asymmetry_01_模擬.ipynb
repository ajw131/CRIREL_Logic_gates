# Cout

```python
# single and test time 
bo = 0
wee = 5
wei = 14 # 14
wie = 7
wii = 7
wo = 5
stim = 5
T = 400
a = 1.6
theta = 4
tau = 1
bias = 10
dif = 0.5
delta_e = 0
delta_i = 0

fname = str(bo)+"_"+str(wee).zfill(2)+"_"+str(wei).zfill(2)+"_"+str(wie).zfill(2)+"_"+str(wii).zfill(2)+"_"+str(wo)+"_"+\
        str(stim)+"_"+str(T)+"_"+str(a)+ "_"+str(theta)+ "_"+str(tau)+ "_"+str(bias)+ "_"+str(dif)+ "_"+\
        str(delta_e)+"_"+str(delta_i)

paras = (bo, wee, wei, wie, wii, wo, stim, T, a, theta, tau, bias, dif)

print(f"fname = {fname}")
# ============== parameter =============================
e_bias_list = np.arange(-bias, bias*1.5, dif)
i_bias_list = np.arange(-bias, bias*1.5, dif)
# label_list = np.zeros([len(e_bias_list)*len(i_bias_list)])

logic_FR_all    = np.zeros([len(i_bias_list), len(e_bias_list), 4])   # 四個時段的FR
logic_label_all = np.zeros([len(i_bias_list), len(e_bias_list)])      # logic type



# ====== weight =====
W1 = np.array([[  0,   0,   0,    0,   0],
               [1.5,   0, wee, -wei,   0],
               [1.5, wee,   0,   0, -wei],
               [ -2, wie,   0,   0, -wii],
               [ -2,   0, wie, -wii,   0]])
    
W2 = np.array([[wo, wo]])

# bias
e_bias = -0.5
i_bias = -5

print(f"e_bias={e_bias}, i_bias={i_bias}")
# ====== bias =====
B1 = [0, e_bias, e_bias+delta_e, i_bias, i_bias+delta_i]
B2 = [bo]

# ============== simulation =============================
data_temp, logic_FR, logic_label = crirel.simulation_crirel_data(paras, e_bias, i_bias, W1, W2, B1, B2)

# num+=1
clear_output()
    
data_df = pd.DataFrame(data_temp, columns=["t", "Cin", "E1", "E2", "I1", "I2", "EO", "Inp1", "Inp2"])
data_df_c = pd.DataFrame(data_temp, columns=["t", "Cin", "E1", "E2", "I1", "I2", "EO", "Inp1", "Inp2"])

data_df.to_csv('result/adder_data/df_cout.csv', index=False)
print("finish")
```

```python
tit = "Tit"

fig, ax = plt.subplots(3, 1, figsize=(5, 5.5), sharex=True, facecolor="white")
plt.subplots_adjust(hspace=0.2, wspace=0.2, left=0.2)
FZ = 16

# st = fig.suptitle(f"{tit}, e_bias={e_bias},i_bias={i_bias}\n delta=({delta_e},{delta_i})", fontsize=FZ+2)
st = fig.suptitle(f"Cout (AND/OR)", fontsize=FZ+6)

ax[0].plot(data_df["t"], data_df["Cin"], color='#800000', label='Cin')  
ax[0].set_ylim([-0.1, 1.4])

ax[1].plot(data_df["t"], data_df["Inp1"], 'g', label=r'$A$')
ax[1].plot(data_df["t"], data_df["Inp2"], 'm', linewidth=2, dashes=[6, 4], label=r'$B$')
ax[1].set_ylim([-1, 9])


ax[2].plot(data_df["t"], data_df["EO"], 'b', label='Eout')
ax[2].set_ylim([-0.1, 1.5])

ax[2].set_xlabel('time (ms)', fontsize=FZ)
ax[2].text(1*T+100, 1.15, "AND", fontsize=FZ+2, c="r")
ax[2].text(5*T+100, 1.15, "OR", fontsize=FZ+2, c="r")

ax[0].text( 700, 1.1, 0, fontsize=FZ, c="k")
ax[0].text(2300, 1.1, 1, fontsize=FZ, c="k")


inp_list = ["11", "10", "01", "00"]
for j in range(4):
    ax[1].text(50+400*j, 6, inp_list[j], fontsize=FZ, c="k")
    ax[1].text(1650+400*j, 6, inp_list[j], fontsize=FZ, c="k")

cols = [r'$r_{Compare}$', 'Input (nA)', r'$r_{Compare}$']
for i, col in enumerate(cols):
    ax[i].set_ylabel(col , fontsize=FZ)
    ax[i].axvline(x=4*T, ymin=0, ymax=6, color = 'r', linestyle = '--', alpha=0.5)

ax[1].legend(loc='center right', bbox_to_anchor=(1.26, 0.5), handlelength=1, fontsize=FZ)

ax[2].axhline(y=0.5, xmin=0, xmax=3000, color = 'k', linestyle = '--', alpha=0.3)

ax[0].set_yticks([0,1], ["0","5"])
ax[1].set_yticks([0,5], ["0","5"])
ax[2].set_yticks([0,0.5,1], ["0","0.5","1"])
plt.subplots_adjust(top=0.9)


plt.savefig('result/adder_plot/Cout.png', dpi=200)
```
